```python
#importiamo la libreria pandas
import pandas as pd
```

```python
#inseriamo il file (raw) che voliamo visualizzare
df = pd.read_csv("https://raw.githubusercontent.com/Krieg2065/Python/main/titanic")
df
```

```python
#visualizziamo le prime 10 righe
df.head(10)
```

```python
#visualizziamo le ultime 10 righe
df.Fax(10)
```

```python
#visualizzare solo alcune colonne
df[["Name", "Sex"]]
```

```python
#visualizzare solo una colonna in modo grezzo
df["Name"]
```

```python
#troviamo tutte le persone con l'età 19
df['Age'].max()
```

```python
#serie
eta19 = df.Age == 19
df[eta19]
type(eta19)
```

```python
#tutte le persone di età maggiore di 35 anni
df[df.Age > 35]
```

```python
#tutte le persone di età compresa tra i 35/40 anni ordinati in modo crescente
df[(df.Age > 35) & (df.Age < 40)]
```

```python
#tutte le persone di età compresa tra i 35/40 anni ordinati in modo decrescente
df[(df.Age > 35) & (df.Age < 40)].sort_values("Age", ascending = False)
```

```python
#tutte le persone di età compresa tra i 35/40 anni ordinati in modo decrescente compreso il sesso
df[(df.Age > 35) & (df.Age < 40)].sort_values(["Age", "Sex"], ascending = False)
```

```python
#tutte le persone di età compresa tra i 35/40 anni ordinati in modo decrescente compreso il sesso e visualizziamo anche il nome e l'età
df[(df.Age > 35) & (df.Age < 40)].sort_values(["Age", "Sex"], ascending = False)[["Name", "Sex", "Age"]]
```

```python
#età media
df.Age.mean()
```

```python
#età media dei maschi
df[df.Sex == "female"]["Age"].mean()
```

```python
df.groupby("Sex").mean()[["Age"]]
```
